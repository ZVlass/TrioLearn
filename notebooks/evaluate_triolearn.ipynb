```python
#  Run the notebook that defines `recommend_for_keywords`

%run 04_keyword_recommendations.ipynb
import sys
sys.path.append("..")  # so that utils/ is on the path
from utils.evaluation_metrics import precision_at_k
import pandas as pd
```

```python
# A tiny example—swap in test set / annotations
test_queries = [
    "intro to machine learning",
    "advanced python programming",
    "data visualization with matplotlib"
]

# For each query, a set of global_id's you consider “relevant”
ground_truth_sets = [
    {101, 203, 305},
    {150, 178, 190, 202},
    {310, 315}
]

assert len(test_queries) == len(ground_truth_sets)
```

```python
recommended_lists = []

for q in test_queries:
    df_rec = recommend_for_keywords(q, top_k=5)    # use k=5 or k=10 as you like
    rec_ids = df_rec['global_id'].tolist()         # extract just the IDs
    recommended_lists.append(rec_ids)

# inspect
for q, rec in zip(test_queries, recommended_lists):
    print(f"> Query: {q!r}\n  Top IDs: {rec}\n")

```

```python
prec5 = precision_at_k(recommended_lists, ground_truth_sets, k=5)
print(f"Precision@5 over {len(test_queries)} queries: {prec5:.3f}")
```
